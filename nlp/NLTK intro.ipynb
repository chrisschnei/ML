```python
import nltk
import pandas as pd
import os

nltk.download("gutenberg")
from nltk.corpus import gutenberg
print(" | ".join(gutenberg.fileids()))

nltk.download("inaugural")
from nltk.corpus import inaugural
print(" / ".join(inaugural.fileids()))

nltk.download("reuters")
from nltk.corpus import reuters
print(" | ".join(reuters.categories()))

print(reuters.fileids()[:5], "\n")
print(reuters.fileids()[-5:], "\n")

reuters.categories("training/999")
reuters.fileids("interest")[:5]

files = gutenberg.fileids()
" | ".join(files)

nltk.download('punkt')
# corpus analysis tools
n_chars = [len(gutenberg.raw(f)) for f in files]
n_words = [len(gutenberg.words(f)) for f in files]
n_sents = [len(gutenberg.sents(f)) for f in files]
n_vocab = [len(set(w.lower() for w in gutenberg.words(f))) for f in files]

gutenberg_df = pd.DataFrame({"file_id": gutenberg.fileids(),
                             "n_chars": n_chars,
                             "n_words": n_words,
                             "avg_word_len": pd.Series(n_chars)/pd.Series(n_words),
                             "n_sents": n_sents,
                             "avg_sent_len": pd.Series(n_words)/pd.Series(n_sents),
                             "n_vocab": n_vocab,
                             "type_token_ratio": (pd.Series(n_vocab)/pd.Series(n_words))*100
                             })

gutenberg_df.sort_values(by="n_words", ascending=False).round(2)

gutenberg_df.sort_values(by="type_token_ratio", ascending=False).round(2)

# stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords_en = stopwords.words("english")
print("Num. Englische StopwÃ¶rter: ", len(stopwords_en))
print("Englische StopwÃ¶rter: ", ", ".join(stopwords_en))

from nltk.corpus import brown
nltk.download('brown')
brown_categories = brown.categories()
print(f"Genres in Brown corpus: {', '.join(brown_categories)}")

adventure_text = brown.words(categories="adventure")
freq_dist = nltk.FreqDist(w.lower() for w in adventure_text)

sorted_freq_dist = {k: v for k, v in sorted(freq_dist.items(), key=lambda item: item[1], reverse=True)}

for k,v in sorted_freq_dist.items():
    if k not in stopwords_en:
        print(f"{k}: {v}")

conditionals = []
for genre in brown.categories():
    for word in brown.words(categories=genre):
        conditionals.append((genre,word))

conditionals[:10]

cfd_genre = nltk.ConditionalFreqDist(conditionals)
cfd_genre.tabulate(conditions=brown_categories, samples=stopwords_en[:10])

print("Deutsche StopwÃ¶rter", " ,".join(stopwords.words("german")))
print("\nSprachen: ", os.listdir("/root/nltk_data/corpora/stopwords"))
```
